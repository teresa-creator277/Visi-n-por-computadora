# Análsis de ventas con NumPy

Supongamos que tienes datos de ventas de una tienda en diferentes meses y quieres analizarlos para obtener información clave, como el total de ventas, el promedio de ventas por mes y los meses con mejor desempeño.

## Paso 1. Importar NumPy y definir los datos


```python
import numpy as np

# Datos de ventas por cada mes
ventas = np.array([2300, 2500, 2600, 2100, 2900, 3100, 2800, 2700, 3000, 3300, 3500, 3700])
# Meses del año
meses = np.array(["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
                  "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"])
print(ventas)
print(meses)
```

## Paso 2. Análisis de ventas con Numpy

Aplicaremos funciones clave para extraer información útil

### 1. Calcular el total y promedio de ventas al año.

```python
# Obtener el total de las ventas y el promedio, imprimirlas
total_ventas = np.sum(ventas)
promedio_ventas = np.mean(ventas)

print(f"Total de ventas del año: {total_ventas}")
print(f"Promedio de ventas mensuales: {promedio_ventas:.2f}")
```

### 2. Identificar el mes con más y menos ventas

```python
# Mes con ventas maximas
indice_max = np.argmax(ventas)
mes_max = meses[indice_max]
print(f"Mes con ventas máximas: {mes_max} ({ventas[indice_max]})")

# Mes con ventas minimas
indice_min = np.argmin(ventas)
mes_min = meses[indice_min]
print(f"Mes con ventas mínimas: {mes_min} ({ventas[indice_min]})")
```

### 3. Determinar los meses con ventas superiores al promedio

```python
# Meses con Ventas mayores al promedio
meses_mayores_promedio = meses[ventas > promedio_ventas]
ventas_mayores_promedio = ventas[ventas > promedio_ventas]

print("Meses con ventas mayores al promedio:")
for mes, venta in zip(meses_mayores_promedio, ventas_mayores_promedio):
    print(f"{mes}: {venta}")
```

## 4. Aplicar un incremento del 10% en las ventas

```python
ventas_incrementadas = ventas * 1.10
print(f"Ventas proyectadas para el próximo año {ventas_incrementadas}")
```

## 5. Gráficar las ventas actuales y el incremento para el próximo año

```python
# importar la libreria para gráficar matplotlib
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(meses, ventas, marker='o', label='2024', linestyle='dashed', color='b')
plt.plot(meses, ventas_incrementadas, marker='s', label='2025', linestyle='solid', color='r')

# Agregar las etiquetas
plt.xlabel("Meses")
plt.ylabel("Ventas")
plt.title("Comparación de Ventas Actuales vs. Proyectadas")
plt.legend()
plt.xticks(rotation=45)  # Rotar etiquetas de meses para mejor visualización
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

## 7. Detectar anomalias con un gráfico de caja (Boxplot)

```python
import seaborn as sns

# Crear la gráfica de caja
plt.figure(figsize=(6, 5))
sns.boxplot(y=ventas, color='skyblue')

# Etiquetas y título
plt.ylabel("Ventas")
plt.title("Detección de Anomalías en Ventas (Distribución de ventas)")

# Mostrar la gráfica
plt.show()
```
