```python
import numpy as np
import tensorflow as tf
```

```python
# 1. Definir las entradas (MXN) y salidas (USD) - Dataset de 1000 elementos
np.random.seed(42)
x_mxn = np.random.uniform(5000, 50000, 1000)  # 1000 valores entre 5,000 y 50,000 MXN
y_usd = x_mxn / 17.8  # Conversión exacta a USD
```

```python
# Verificar el tamaño (deben ser iguales)
print("Tamaño de entradas:", x_mxn.shape, "Tamaño de salidas:", y_usd.shape)
```

```python
# 2. Normalizar los datos
x_mxn_normalized = x_mxn / 10000  # Escalar MXN
y_usd_normalized = y_usd / 1000   # Escalar USD
```

```python
# 3. Definir hiperparámetros
learning_rate = 0.05
epochs = 1000
```

```python
# 4. Diseñar red neuronal (similar a tu estructura)
capa1 = tf.keras.layers.Dense(units=3, input_shape=[1])
capa2 = tf.keras.layers.Dense(units=3)
capa3 = tf.keras.layers.Dense(units=1)
```

```python
# 5. Crear modelo
modelo = tf.keras.Sequential([capa1, capa2, capa3])
```

```python
# 6. Compilar el modelo
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate),
    loss='mean_squared_error',
    metrics=['mae']
)
```

```python
# 7. Entrenar el modelo
print("Comenzando entrenamiento...")
historial = modelo.fit(x_mxn_normalized, y_usd_normalized, epochs=epochs, verbose=False)
print("Modelo entrenado!")
```

```python
# 8. Obtener pesos y bias ajustados
print("\nPesos y bias de la última capa:")
print(capa3.get_weights())
```

```python
# 9. Evaluar el modelo (accuracy personalizado)
def calcular_accuracy(model, X, y, margen=0.01):
    predicciones = model.predict(X).flatten()
    correctos = np.sum(np.abs(predicciones - y) < margen)
    return (correctos / len(y)) * 100

train_acc = calcular_accuracy(modelo, x_mxn_normalized, y_usd_normalized)
print(f"\nAccuracy en entrenamiento: {train_acc:.2f}%")
```

```python
# 10. Probar con datos nuevos
test_mxn = np.array([17800, 35600]) / 10000  # Normalizar
predicciones = modelo.predict(test_mxn) * 1000  # Desnormalizar

print("\nPrueba con datos nuevos:")
for mxn, usd in zip([17800, 35600], predicciones.flatten()):
    print(f"{mxn} MXN → {usd:.2f} USD (Valor real: {mxn/17.8:.2f} USD)")
```

```python
# 11. Guardar el modelo
modelo.save('conversion_mxn_usd.keras')
print("\nModelo guardado como 'conversion_mxn_usd.keras'")
```
